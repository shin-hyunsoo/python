#### 아래 셀에 variables, neighbors, domains, constraints를 구현하시오. (csp.py의 Sudoku 클래스 참고)

```python
## TODO

# variables = NotImplemented

# neighbors = NotImplemented

# domains = NotImplemented

# constraints = NotImplemented
```

```python
variables = ['A','B','C','D','E','F','G']
neighbors = {
    'A':['B','C','D','G'],
    'B':['A','F'],
    'C':['A','D','E','G'],
    'D':['A','C','F','G'],
    'E':['C'],
    'F':['B','D'],
    'G':['A','C','D']
    }
domains = {'A':[2,4],
          'B':[4],
          'C':[1,3],
           'D':[3,4],
           'E':[3],
           'F':[3,4],
           'G':[3]
          }
constraints = different_values_constraint
```

#### solution 값을 출력하여 A, B, C, D, E, F, G 값을 확인하시오. (아래 셀은 수정하지 마십시오.)

```python
problem = CSP(variables, domains, neighbors, constraints)

solution = backtracking_search(problem, inference=forward_checking)

print(solution)
```
